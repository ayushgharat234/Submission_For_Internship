```python
# Importing necessary libraries
import pandas as pd
import os
import zipfile
```

```python
# Loadibng the zipped data in a new directory, same as the location of this notebook
curr_directory = os.getcwd()

# Create a folder/path of the data to be stored
data_directory = os.path.join(curr_directory,'data')
os.makedirs(data_directory,exist_ok=True)

# Downloaded file path
zip_file_path = "C:\\Users\\Gharat\\Downloads\\movie_data.zip"

# Unzip the file
with zipfile.ZipFile(zip_file_path,'r') as zip_key:
    zip_key.extractall(data_directory)

# holding the data(temporary)
data_dictionary = {}

# loading the CSV files only into the dictionary
for filename in os.listdir(data_directory):
    if filename.endswith('.csv'):  
        path = os.path.join(data_directory,filename)

        data = pd.read_csv(path)
        data_dictionary[filename] = data

# Imported data info
print("DataFrames imported from the following files:")
print(list(data_dictionary.keys()))
```

```python
# Data loading into separate data frames 
for filename, data in data_dictionary.items():
    filename = filename.replace(".csv","") + "_data"

    globals()[filename] = data
```

```python
import requests

# the API Ley
API_KEY = '15f38c1c9c9a4e3c44dd2f5f8845a7ee'

# Scrapper function
def fetch_tmdb_rating(tmdb_id):
    URL = f"https://api.themoviedb.org/3/movie/{tmdb_id}?api_key={API_KEY}&language=en-US"
    response = requests.get(URL)
    if response.status_code == 200:
        data = response.json()
        rating = data.get('vote_average', None)
        return rating
    else:
        return None

# Applying the function
links_data['rating'] = links_data['tmdbId'].apply(fetch_tmdb_rating)
```
